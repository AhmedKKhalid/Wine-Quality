# Preprocessing Data

```python
def rank(x):
    if x in range(0,6):
        return 0 # Bad
    elif x >5:
        return 1 # Good
```

```python
df['quality']=df['quality'].apply(lambda x:rank(x))
```

```python
df['quality'].value_counts().plot.bar()
```

# Building Model

```python
X=df.drop('quality',axis=1)
Y=df['quality']
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2)
```

### Hyper parameter 

```python
collection = {
    
        'random_forest':{
           'model':RandomForestClassifier(),
            'params':{
                'n_estimators':[1,5,10,30,40,50]
            } 
        },
        'logistic_reg':{
            'model':LogisticRegression(solver='liblinear',multi_class='auto'),
            'params':{
                'C':[1,5,10],
            }
        },
        'Knn':{
            'model':KNeighborsClassifier(),
            'params':{
                'n_neighbors':[1,3,5,10],
            }
        },
        
    }
```

```python
for name,data in collection.items():
    clf=GridSearchCV(data['model'],data['params'],cv=5,return_train_score=False)
    clf.fit(x_train,y_train)
    print("{},{} =>{} ".format(name,clf.best_params_,clf.best_score_))
```

### Random Forest Model

```python
model1=RandomForestClassifier(n_estimators=40)
model1.fit(x_train,y_train)
model1.score(x_test,y_test)
```

# Model Evaluation

```python
y_pred=model1.predict(x_test)
print(classification_report(y_test,y_pred))
```

```python
cn=confusion_matrix(y_test,y_pred)
sns.heatmap(cn,annot=True)
```

```python
ac = accuracy_score(y_test,model1.predict(x_test))
print('Accuracy is: ',ac)
```

```python
def DistributionPlot(RedFunction, BlueFunction, RedName, BlueName, Title):
    width = 12
    height = 10
    plt.figure(figsize=(width, height))

    ax1 = sns.distplot(RedFunction, hist=False, color="r", label=RedName)
    ax2 = sns.distplot(BlueFunction, hist=False, color="b", label=BlueName, ax=ax1)

    plt.title(Title)
    plt.legend()
    plt.show()
    plt.close()
```

```python
DistributionPlot(y_test, y_pred, "Actual Values (Train)", "Predicted Values (Train)", "accuracy dist")
```
